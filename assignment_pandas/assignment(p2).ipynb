```python
def add_weekday(df):
    df['Date'] = pd.to_datetime(df["Date"])
    df['Weekday'] = df['Date'].dt.strftime('%A')

data = {"Date":['2023-07-20','2023-07-21','2023-07-22']}
df = pd.DataFrame(data)
add_weekday(df)
print(df)
```

*ques 12*

```python
def select_dates(df):
    df['Date'] = pd.to_datetime(df['Date'])
    val = (df['Date']>='2023-01-01') & (df['Date'] <= '2023-01-31')
    selected_rows = df[val]
    return selected_rows

data = {'Date': ['2023-01-15', '2023-01-25', '2023-02-10', '2023-01-05','2023-01-30']}
df = pd.DataFrame(data)
selected_dates = select_dates(df)
print(selected_dates)
```
